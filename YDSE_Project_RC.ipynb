```python
import numpy as np
import matplotlib.pyplot as plt

d = float(input("Enter d (in mm): ")) * 1e-3
lam_nm = float(input("Enter lambda (in nm): "))
lam = lam_nm * 1e-9
D = float(input("Enter D (in m): "))
psi = np.radians(float(input("Enter psi (in degrees): ")))

def G(n):
    u = (d * np.cos(psi)) / (lam * n)
    return 1 - (u**2)

x = np.linspace(-1, 1, 1000)
y = np.linspace(-1, 1, 1000)
X, Y = np.meshgrid(x, y)

def c(r):
    if 380 <= r < 445:
        return 'violet'
    elif 445 <= r < 460:
        return 'indigo'
    elif 460 <= r < 495:
        return 'blue'
    elif 495 <= r < 570:
        return 'green'
    elif 570 <= r < 590:
        return 'yellow'
    elif 590 <= r < 620:
        return 'orange'
    elif 620 <= r < 750:
        return 'red'
    else:
        return 'white'

color = c(lam_nm)

for i in range(2, 15):
    F = X**2 + G(i**3)*(Y**2) + 2*D*Y*np.sin(psi) + D**2
    plt.contour(X, Y, F, levels=[0], colors=color)

plt.gca().set_aspect('equal')
plt.gca().set_facecolor('black')
plt.show()
```
